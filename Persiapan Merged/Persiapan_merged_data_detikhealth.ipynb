```python
import pandas as pd
import dateparser

# Fungsi untuk mengonversi format tanggal
def convert_date_format(date_str):
    # Mengurai tanggal menggunakan dateparser
    date = dateparser.parse(date_str)
    # Memeriksa apakah parsing berhasil sebelum melakukan konversi
    return date.strftime("%m-%d-%Y") if date else None

# Terapkan fungsi pada kolom Tanggal di DataFrame Anda
df['date'] = df['date'].apply(convert_date_format)
print(df)
```
